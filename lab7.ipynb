```python
import nltk

# Lấy danh sách tất cả các corpus có sẵn trong NLTK Data
# Lưu ý: Không có một hàm trực tiếp để liệt kê TẤT CẢ các corpus một cách tổng quát
# Tuy nhiên, chúng ta có thể liệt kê các corpus phổ biến hoặc các corpus đã được tải.
# Ví dụ, liệt kê các corpus trong nltk.corpus

print("Các corpus phổ biến trong NLTK (từ nltk.corpus):")

# Cách 1: Liệt kê các thuộc tính của nltk.corpus và kiểm tra nếu chúng có phương thức 'fileids'
corpus_names = []
for attr in dir(nltk.corpus):
    if not attr.startswith('_'):
        try:
            corpus_obj = getattr(nltk.corpus, attr)
            if hasattr(corpus_obj, 'fileids'):
                corpus_names.append(attr)
        except LookupError:
            # Bỏ qua các corpus không tìm thấy hoặc chưa được tải đầy đủ
            pass
        except Exception as e:
            # Xử lý các lỗi khác có thể xảy ra khi truy cập corpus
            pass

print("Danh sách các CorpusReader:\n", corpus_names)

# Ví dụ cụ thể với một số corpus đã được tải
print("\nCác corpus đã tải (ví dụ):")
print("- Gutenberg: ", nltk.corpus.gutenberg.fileids()[:5], "...") # Chỉ in 5 file đầu
try:
    print("- Brown: ", nltk.corpus.brown.fileids()[:5], "...")
except LookupError:
    print("- Brown: [Corpus 'brown' chưa thể truy cập được hoặc không tìm thấy sau khi tải xuống]...")
print("- Movie Reviews: ", nltk.corpus.movie_reviews.fileids()[:5], "...")
print("- Names: ", nltk.corpus.names.fileids())
print("- Stopwords: ", nltk.corpus.stopwords.fileids())
```

### 4. Loại bỏ các stopword từ một văn bản đã cho

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = "Đây là một ví dụ về câu có chứa các từ dừng tiếng Anh và tiếng Việt."
english_stopwords = set(stopwords.words('english'))
vietnamese_stopwords = ["là", "một", "về", "và", "có"]

# Kết hợp stopwords của nhiều ngôn ngữ nếu cần, hoặc chỉ sử dụng một ngôn ngữ
all_stopwords = english_stopwords.union(set(vietnamese_stopwords))

# Tokenize văn bản
words = word_tokenize(text.lower())

# Lọc bỏ stopwords
filtered_words = [word for word in words if word not in all_stopwords]

print("Văn bản gốc:", text)
print("Các từ sau khi loại bỏ stopwords:", filtered_words)
```

### 7. Tìm tập hợp các từ đồng nghĩa và trái nghĩa của một từ nào đó

```python
import nltk
from nltk.corpus import wordnet

word = "good"

synonyms = []
antonyms = []

for syn in wordnet.synsets(word):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())

print(f"Từ gốc: {word}")
print(f"Các từ đồng nghĩa: {set(synonyms)}")
print(f"Các từ trái nghĩa: {set(antonyms)}")

word_example_2 = "bright"
synonyms_2 = []
antonyms_2 = []

for syn in wordnet.synsets(word_example_2):
    for l in syn.lemmas():
        synonyms_2.append(l.name())
        if l.antonyms():
            antonyms_2.append(l.antonyms()[0].name())

print(f"\nTừ gốc: {word_example_2}")
print(f"Các từ đồng nghĩa: {set(synonyms_2)}")
print(f"Các từ trái nghĩa: {set(antonyms_2)}")
```

### 9. So sánh sự giống nhau của hai danh từ đã cho

```python
import nltk
from nltk.corpus import wordnet

# Hàm để lấy synset của danh từ
def get_noun_synset(word):
    synsets = wordnet.synsets(word, pos=wordnet.NOUN)
    return synsets[0] if synsets else None

word1 = "cat"
word2 = "dog"
word3 = "car"

syn1 = get_noun_synset(word1)
syn2 = get_noun_synset(word2)
syn3 = get_noun_synset(word3)

print(f"Sự giống nhau giữa '{word1}' và '{word2}':")
if syn1 and syn2:
    print(f"  Đường đi ngắn nhất: {syn1.path_similarity(syn2):.3f}")
    print(f"  Leacock-Chodorow: {syn1.lch_similarity(syn2):.3f}")
    # Wu-Palmer similarity requires a common root. The higher the value, the more similar the words.
    print(f"  Wu-Palmer: {syn1.wup_similarity(syn2):.3f}")
else:
    print("  Không thể tính toán sự giống nhau.")

print(f"\nSự giống nhau giữa '{word1}' và '{word3}':")
if syn1 and syn3:
    print(f"  Đường đi ngắn nhất: {syn1.path_similarity(syn3):.3f}")
    print(f"  Leacock-Chodorow: {syn1.lch_similarity(syn3):.3f}")
    print(f"  Wu-Palmer: {syn1.wup_similarity(syn3):.3f}")
else:
    print("  Không thể tính toán sự giống nhau.")
```

### 10. So sánh sự giống nhau của hai động từ đã cho

```python
import nltk
from nltk.corpus import wordnet

# Hàm để lấy synset của động từ
def get_verb_synset(word):
    synsets = wordnet.synsets(word, pos=wordnet.VERB)
    return synsets[0] if synsets else None

verb1 = "run"
verb2 = "walk"
verb3 = "eat"

syn_v1 = get_verb_synset(verb1)
syn_v2 = get_verb_synset(verb2)
syn_v3 = get_verb_synset(verb3)

print(f"Sự giống nhau giữa '{verb1}' và '{verb2}':")
if syn_v1 and syn_v2:
    print(f"  Đường đi ngắn nhất: {syn_v1.path_similarity(syn_v2):.3f}")
    print(f"  Leacock-Chodorow: {syn_v1.lch_similarity(syn_v2):.3f}")
    print(f"  Wu-Palmer: {syn_v1.wup_similarity(syn_v2):.3f}")
else:
    print("  Không thể tính toán sự giống nhau.")

print(f"\nSự giống nhau giữa '{verb1}' và '{verb3}':")
if syn_v1 and syn_v3:
    print(f"  Đường đi ngắn nhất: {syn_v1.path_similarity(syn_v3):.3f}")
    print(f"  Leacock-Chodorow: {syn_v1.lch_similarity(syn_v3):.3f}")
    print(f"  Wu-Palmer: {syn_v1.wup_similarity(syn_v3):.3f}")
else:
    print("  Không thể tính toán sự giống nhau.")
```

### 11. Tìm số lượng tên nam và nữ trong các tên kho ngữ liệu. In tên 10 nam và nữ đầu tiên.

```python
import nltk
from nltk.corpus import names

# Tải gói 'names' nếu chưa có
# nltk.download('names') # Đã tải ở đầu notebook

male_names = names.words('male.txt')
female_names = names.words('female.txt')

print(f"Tổng số tên nam: {len(male_names)}")
print(f"10 tên nam đầu tiên: {male_names[:10]}")

print(f"\nTổng số tên nữ: {len(female_names)}")
print(f"10 tên nữ đầu tiên: {female_names[:10]}")
```

### 12. In 15 kết hợp ngẫu nhiên đầu tiên được gắn nhãn nam và được gắn nhãn tên nữ từ kho tên.

```python
import nltk
import random
from nltk.corpus import names

# Tải gói 'names' nếu chưa có
# nltk.download('names') # Đã tải ở đầu notebook

male_names = names.words('male.txt')
female_names = names.words('female.txt')

# Tạo danh sách các cặp (tên, nhãn)
labeled_names = ([(name, 'male') for name in male_names] +
                 [(name, 'female') for name in female_names])

# Xáo trộn danh sách
random.shuffle(labeled_names)

print("15 kết hợp ngẫu nhiên đầu tiên (tên và nhãn):")
for i in range(15):
    print(labeled_names[i])
```
